```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from mplsoccer.pitch import Pitch
```

```python
passes = pd.read_csv("Passes.csv", delimiter = ";")
```

```python
passes.head()
```

```python
passes["x"] = passes["x"]*1.2
passes["y"] = passes["y"]*.8
passes["endX"] = passes["endX"]*1.2
passes["endY"] = passes["endY"]*.8
```

```python
passes.head()
```

```python
shots = pd.read_csv("Shots.csv", delimiter = ";")
```

```python
shots
```

```python
fig, ax = plt.subplots(figsize = (13, 8))
fig.set_facecolor("#22312b")
ax.patch.set_facecolor("#22312b")

pitch = Pitch(pitch_type = "statsbomb",
              ##orientation='horizontal',
              pitch_color = "#22312b",
              line_color = "#c7d5cc")
              ##figsize = (10, 7),
              ##constrained_layout = True, 
              ##tight_layout = False)

pitch.draw(ax=ax)
##plt.gca.invert_yaxis()

for i in range(len(passes["x"])):
    if passes["outcome"][i] == "Successful":
        plt.plot((passes["x"][i],passes["endX"][i]), 
                 (passes["y"][i],passes["endY"][i]),
                color = "yellow")
        
    if passes["outcome"][i] == "Unsuccessful":
        plt.plot((passes["x"][i],passes["endX"][i]), 
                 (passes["y"][i],passes["endY"][i]),
                color = "red")

plt.title('Messi Pass Map vs Real Betis',color='white',size=20)
plt.show()
```
