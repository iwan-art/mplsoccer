```python
from mplsoccer.pitch import Pitch, VerticalPitch
```

### Basic Pitch

```python
pitch = Pitch(pitch_color = "grass",
              line_color = "white",
              stripe = True)
fig, ax = pitch.draw()
```

```python
pitch = Pitch(half = False)
fig, ax = pitch.draw()
```

### Vertical Full

```python
pitch = VerticalPitch(half = False)
fig, ax = pitch.draw()
```

### Horizontal Half

```python
pitch = Pitch(half = True,
             corner_arcs = True)
fig, ax = pitch.draw()
```

### Line Style

```python
pitch = Pitch(linestyle = "--",
             linewidth = 1,
             goal_linestyle = "-")
fig, ax = pitch.draw()
```
